```python
from langchain_community.vectorstores import Chroma
from tuyabot_llm import AbsolutePaths
```

# Embeddings

```python
from langchain.embeddings import SentenceTransformerEmbeddings

embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

# Load original documents

```python
current_dir = AbsolutePaths().get_abs_path_folder('raw')
db_dir = os.path.join(current_dir, "tuya_collection")
persistent_directory_original = os.path.join(db_dir, "chroma_tuya_collection")

db = Chroma(persist_directory=persistent_directory_original, embedding_function=embeddings)
```

```python
# Step 5: Query the vector store
retriever = db.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 10},
)

# Define the user's question
query = "Tuya"

# Retrieve relevant documents based on the query
documents = retriever.invoke(query)

# Display the relevant results with metadata
print("\n--- Relevant Documents ---")
for i, doc in enumerate(documents, 1):
    print(f"Document {i}:\n{doc.page_content}\n")
    if doc.metadata:
        print(f"Source: {doc.metadata.get('source', 'Unknown')}\n")
```

```python
for document in documents:
    document.page_content = document.page_content.strip()
    document.page_content = document.page_content.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ')
    document.page_content = ' '.join(document.page_content.split())
```

# Chunk files

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=40,
    length_function=len
)

docs = text_splitter.split_documents(documents)
```

# Save in Vector DB

```python
current_dir = AbsolutePaths().get_abs_path_folder('raw')
db_dir = os.path.join(current_dir, "tuya_collection")
persistent_directory_clean = os.path.join(db_dir, "chroma_tuya_collection_cleaned")
```

```python
if not os.path.exists(persistent_directory_clean):
    print(f"\n--- Creating vector store in {persistent_directory_clean} ---")
    db = Chroma.from_documents(docs, embeddings, persist_directory=persistent_directory_clean)
    print(f"--- Finished creating vector store in {persistent_directory_clean} ---")
else:
    print(f"Vector store {persistent_directory_clean} already exists. No need to initialize.")
    db = Chroma(persist_directory=persistent_directory_clean, embedding_function=embeddings)
```
